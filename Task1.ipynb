```python
def stringCounter(strings):
    count = 0
    for i in strings:
        if len(i)>=2 and i[0][0]== i[-1][-1]:
            count += 1
    return count
```

# Task 4

```python
def SortStrings(strings):
    n= strings.sort()
    final=[]
    for i in n:
        if i[0][0] == 'x' or i[0][0] == 'X':
            final.append(i)
    for i in n:
        if i not in final:
            final.append(i)
    return final
        
```

# Task 5

```python
def sortTuple(tuples,m):
    def last(n):
        return n[m]
    return sorted(tuples, key = last)
```

# Task 6

```python
def RemoveAdjacent(lists):
    final=[]
    for i,num in enumerate(lists):
        if lists[i+1]!=num:
            final.append(num)
    return final
```
